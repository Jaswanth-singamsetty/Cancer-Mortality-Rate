#Checking & Dropping columns with help of correlation 

```python
import numpy as np
```

```python
def remove_Highly_Correlated(df, bar=0.9):
  # Creates correlation matrix
  corr = df.corr()

  # Set Up Mask To Hide Upper Triangle
  mask = np.triu(np.ones_like(corr, dtype=bool))
  tri_df = corr.mask(mask)

  # Finding features with correlation value more than specified threshold value (bar=0.9)
  highly_cor_col = [col for col in tri_df.columns if any(tri_df[col] > bar )]
  print("length of highly correlated columns",len(highly_cor_col))

  # Drop the highly correlated columns
  reduced_df = df.drop(highly_cor_col, axis = 1)
  print("shape of data",df.shape,"shape of reduced data",reduced_df.shape)
  return reduced_df
```

```python
cleaned_data = remove_Highly_Correlated(df)
```

```python
cleaned_data.head()
```

#Scaling the features 

```python
from sklearn.preprocessing import StandardScaler
features = cleaned_data.iloc[:,cleaned_data.columns!="TARGET_deathRate"] 
StandardScaler().fit(features)
features = StandardScaler().fit_transform(features)
```

#splitting the data into training and testing

```python
x= features   #Features
y= cleaned_data["TARGET_deathRate"]  #Target 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=45)
```

#Training the model

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print("Accuracy=",100-(np.mean((np.abs(y_test-y_pred)/y_test)))*100)
```
